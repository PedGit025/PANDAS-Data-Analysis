# Check if songs are similar using UNION

```python
def get_similarity(song_a, song_b):
    lyrics_song_a = set(song_a.replace('\n', ' ').split(' '))
    lyrics_song_b = set(song_b.replace('\n', ' ').split(' '))

    union_data = lyrics_song_a.union(lyrics_song_b)
    intersection_data = lyrics_song_a.intersection(lyrics_song_b)

    union_total = len(union_data)
    intersection_total = len(intersection_data)

    similarity = intersection_total/union_total
    
    return similarity
```

```python
result_q1 = pd.DataFrame(columns = ['Kendrinck Songs',"Taylor's Lover",'Similarity%'])

for each in df_compiled_data['Title'].unique():
    
    similarity = get_similarity(Lover.lower(),
    df_compiled_data[df_compiled_data['Title'] == each]['Lyrics'].values[0].lower())
    
    results = {
        'Kendrinck Songs':each,
        "Taylor's Lover":Lover,
        'Similarity%': round(similarity * 100,2)
    }
    
    result_q1 = result_q1.append(results, ignore_index=True)
```
