# Analyze the Data

Here we calculate for the total number of UNIQUE words used by Harry per song. Provide insights about harry's discography. What songs stand out in terms of total word count? If you're Harry's data manager, how can you make use of this data? What other metrics can you analyze alongside total words to make it more valuable.

Here we also show the scatterplot of Pyongs (Genius' equivalent of Heart react, which we can use as a proxy variable for interest) and Unique word count, is there any pattern that you can see? What can Harry change if he wants to improve people's interest with his songs?


## Based on the data provided, the data scientist can conclude that;

The more unique words the higher the likes or pyong counts. 

As for other insigths we can say that the songs adore you, cinema, matilda, treat the people Wi, satellite, latenight talk, To be so lonely and sunflower are the most hits for the album. 

i'll advise the manager to tell harry to pattern his next album or songs based on those songs the mean based on the Tone/harmony combined with a lot of unique words.

```python
df_harry_combined['Total Words'] = [len(i) for i in df_harry_combined['song.lyrics'].str.split()]
df_harry_combined['Total Unique Words'] = [len(np.unique(i)) for i in df_harry_combined['song.lyrics'].str.split()]
df_harry_word_stats = df_harry_combined[['song.release_date_components.year','song.title','song.pyongs_count', 'Total Words', 'Total Unique Words']].sort_values(by='Total Words', ascending=False)
df_harry_word_stats['Repetition Index'] = df_harry_word_stats['Total Unique Words']/ df_harry_word_stats['Total Words']
```

```python
plt.figure(figsize=(15,5))
plt.title("Harry Styles' Total UNIQUE Words Used per Song", fontsize=20)
plt.bar(df_harry_word_stats['song.title'].str[:15], df_harry_word_stats['Total Words'])
plt.axhline(df_harry_word_stats['Total Words'].mean(), color='C1', linestyle='--', label='Mean ' + str(df_harry_word_stats['Total Words'].mean()))
plt.xticks(rotation=50)
plt.legend();
```

```python

plt.figure(figsize=(10,10))
plt.title("Scatterplot of Unique words And Pyongs", fontsize=20)

plt.scatter(df_harry_word_stats['Total Unique Words'], 
            df_harry_word_stats['song.pyongs_count'], 
            label = 'Correlation Value = ' + str(np.corrcoef(df_harry_word_stats['Total Unique Words'],df_harry_word_stats['song.pyongs_count'])[0][1])[:5])



plt.legend(fontsize=13)
plt.xlabel('Total Unique Words')
plt.ylabel('Total Pyongs');
```
